```python
def estimate_depth(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply a Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (9, 9), 0)

    # Compute the Laplacian of the blurred image
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)

    # Compute the variance of the Laplacian to estimate depth
    variance = laplacian.var()

    # Depth estimation formula (example)
    # You may need to calibrate this formula based on your specific camera setup
    depth = 1000 / (variance + 1)

    return depth
```
